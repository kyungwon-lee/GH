```python
df = pd.read_csv('AB_NYC_2019.csv', encoding='utf-8') #neighborhood group을 예측을 하는 것이 이번 퀘스트이다.
df.head()
```

```python
our_set = df[df['neighbourhood_group'].isin(['Brooklyn', 'Manhattan'])] #두가지만 저장을 하고 싶어서 our_set에 저장을 했다.
our_set.shape, df.shape
```

### 방 타입에는 어떤 것들이 있나...

```python
set(our_set['room_type'])#중복되는 것을 없애준다.
```

```python
import random

random_seed = random.randint(0,1000)
```

### 자연어를 기계가 알아들을 수 있게 수치로 변환

```python
our_set = our_set.replace('Brooklyn', 0).replace('Manhattan', 1) #string으로 표현하는 기계가 못알아들을 수도 있기에 수치로 바꿔준다.
```

### 룸 타입 역시 자연어로 되어있으므로, "원핫인코딩"을 통해 수치로 변환

```python
#원핫 인코딩이란? --> 초 중 대 이런식의 데이터에게 1,2,3,4 이렇게 수치를 주면 기계는 그럼 초가 대보다 2배 작은건가? 라고 생각할 수 있다.
#그렇기 때문에 column으로 해서 1000이면 초등학교, 0100이면 중학교 이렇게 1과 0으로만 수치 데이터를 배정을 하는 것이다.
x = pd.get_dummies(our_set['room_type']) #원핫 인코딩을 해주는 것이다. room_type으로 나올 수 있는 3가지의 것들을 원핫 인코딩을 해주라는 것이다.
core_cols = our_set[['neighbourhood_group', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month']]
combined = core_cols.merge(x, left_index=True, right_index=True, how='left').reset_index() #옆으로 merge를 하는 것이다. 뒤에 열이 추가되었다.
combined.head()
```

### 불필요한 컬럼 제거, 결측치 제외

```python
final = combined.drop(columns=['index']).dropna() #index가 중복되어 있기 때문에 버렸다.
final.shape
```

```python
### 혹시 아직 결측치가 남아있는지 최종 점검

import missingno as msno #pip install missingno로 설치를 하면 된다. 아래의 그림이 나오는데 결측치가 있는지 없는지를 그림으로 보여준다
msno.matrix(final) #하얀선들 그어지는데 그러면 그부분이 결측되어 있다고 파악할 수 있다. 
```

```python
final.head()
```

### 본격적인 모델링

```python
### 랜덤 시드를 활용하여 트레인, 테스트셋 분리 / 3:1 비율로
#만약 멘하튼의 숙소가 95%라면 절대 올바른 분류를 할 수 없다. 그래서 올바른 test set을 구성을 해줘야 한다. 

from sklearn.model_selection import train_test_split #train set와 test set으로 나눠주는 모듈이다.

X_train, X_test, y_train, y_test = train_test_split(final[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month']], final['neighbourhood_group'], 
                                                    test_size=0.25, random_state=random_seed)
#y train은 train에 사용될 해설지, y test는 test에 활용된 해답지. 75%는 train으로 하고 test는 25%로 하는 것이다. 
```

```python
### 랜덤포레스트 모델 수립, 대충 하이퍼 파라미터 제공
import time #train을 하는데에 얼마나 걸리는지 측정을 하기 위해 import를 한 것이다. 
from sklearn.ensemble import RandomForestClassifier
random_seed = random.randint(0,1000)

rf = RandomForestClassifier(n_estimators=200, oob_score=True, random_state=random_seed) #200은 나무의 개수 200개의 나무를 합쳐서 forest를 만들겠다는 것이다. 
#나무가 많다고 절대 좋아지지 않는다. 최적이 나무 갯수는 몇개일까? 이렇게 생각해보면 된다. 
### 학습 진행
t0 = time.time() #시작할 때의 시간
rf.fit(X_train, y_train) #fitting을 하는 것인데 train을 시킨다고 생각하면 된다. --> 여기가 machine learning의 끝이다. 한줄!
elapsed = time.time()-t0  #끝났을 때의 시간.
print("걸린시간은 {:.2f}초입니다.".format(elapsed)) #우수한 성능.
```

```python
from sklearn.metrics import accuracy_score

### 학습된 모델로 분류 예측
prediction = rf.predict(X_test) #여기서는 y test를 넣으면 안된다. 그럼 해답지를 넣는 셈이 되는 것이다. 
### 채점
accuracy = accuracy_score(y_test, prediction) #y_test는 알려진 정답. prediction은 모델을 바탕으로 도출한 답지.
### 성적 발표
print(f'Mean accuracy score: {100*accuracy:.3}%') #60%가 나온다. 똑같은 모델이다하더라도 조금씩 차이는 있을 것이다. 
```

```python
### 오답/정답 시각화를 위한 간단한 히트맵
#까만색이 틀린 것이다. 이때 유용한 점은 어떤 오답이 주로 나올까를 알아볼 때 히트맵을 쓸 수 있다. 
from sklearn.metrics import confusion_matrix
cm = pd.DataFrame(confusion_matrix(y_test, prediction), columns=['Brooklyn', 'Manhattan'], index=['Brooklyn', 'Manhattan'])
sns.heatmap(cm, annot=True, fmt='d')
```

### SVM (Support Vector Machine)

```python
from sklearn.svm import SVC
random_seed = random.randint(0,1000)

##kernel = 'rbf'라고 하면 커널 트릭을 사용할 수 있다.(초평면을 이용해서 분리하는 것이 불가능한 경우!)
##C파라미터는 여유 변수에 관한 파라미터 (C가 커진다면, 오분류에 대한 벌칙을 강하게 주는 것)
##gamma는 커널 자체에 대한 파라미터로, 감마를 증가시키면 일반적으로 훈련 샘플의 영향력이 증가하고 결정 경계가 더욱 부드러워지게 된다.
```

```python
svm = SVC(kernel = 'rbf', C = 0.8, random_state = random_seed, gamma = 'auto')
```

```python
t0 = time.time() #시작할 때의 시간
svm.fit(X_train, y_train) #fitting을 하는 것인데 train을 시킨다고 생각하면 된다. --> 여기가 machine learning의 끝이다. 한줄!
elapsed = time.time()-t0  #끝났을 때의 시간.
print("걸린시간은 {:.2f}초입니다.".format(elapsed)) #우수한 성능.
```

```python
from sklearn.metrics import accuracy_score

y_pred_svc = svm.predict(X_test)
accuracy = accuracy_score(y_test, y_pred_svc) #y_test는 알려진 정답. prediction은 모델을 바탕으로 도출한 답지.
### 성적 발표
print(f'Mean accuracy score: {100*accuracy:.3}%')
```
