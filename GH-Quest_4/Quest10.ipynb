```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. Confusion matrix & metrics

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

## (1) 데이터 준비

```python
# classification 위한 임의의 데이터 생성 -> X: 2개 feature, y:binary 인 데이터 샘플 100개

##make_classification 함수를 쓰면 데이터를 직접 만들 수 있다. x가 두개라는 것을 n_feature를 지정

X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0)
```

```python
X
```

```python
model = LogisticRegression().fit(X,y)

y_score = model.decision_function(X)  
y_pred = model.predict(X)

# decision_function(): 판별함수 결과값 계산 함수에 x를 넣어준 것이다. 
# predict(): y_score가 양수이면 y=1, 음수이면 y=0으로 분류
```

```python
y
```

```python
y_score
```

```python
y_pred
```

```python
# 데이터프레임 형태로 확인해 보자

df = pd.DataFrame(np.vstack([y_score, y_pred, y]).T,columns=["y_score", "y_pred", "y"])
df = df.sort_values("y_score", ascending=False).reset_index(drop=True)
```

```python
display(df.head(), df.tail())
##맨 오른 쪽의 y가 actual 값이고 y_pred가 예측한 값이다.
```

## (2) confusion matrix 생성

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_predict
```

```python
print(confusion_matrix(y, y_pred)) # TP FN FP TN

##이거를 해석해보면 실제 Y값이 1인데 예측도 1인게 48개 
##실제가 0인데 예측한 것이 1인게 6개
##실제가 1인데 예측이 0인 것이 3개
##0으로 예측한 값과 실제 값도 0인 것이 43개
```

```python
print('accuracy score: ',accuracy_score(y, y_pred) )
print('precision score: ',precision_score(y, y_pred) )
print('recall score: ',recall_score(y, y_pred) )
print('f1 score: ',f1_score(y, y_pred) )
```

```python
print(classification_report(y, y_pred))
##classification_report라는 함수를 쓰면 더 깔끔하게 나타낼 수 있다.
```

```python
# 로지스틱 회귀의 threshold(임계값) 를 0이 아닌 -2로 지정해 보자 
lower_threshold = model.decision_function(X) > -2

df2 = pd.DataFrame(np.vstack([y_score, lower_threshold, y]).T,
                  columns=["y_score", "y_hat", "y"])
df2 = df2.sort_values("y_score", ascending=False).reset_index(drop=True)
```

```python
display(df2[(df2['y_score']> -2.5) & (df2['y_score']<-1.5)])
```

```python
print(confusion_matrix(y, lower_threshold))
```

```python
print(classification_report(y, lower_threshold))
```

# 2. ROC, PR 곡선 그리기

```python
from sklearn.metrics import auc
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_recall_curve
from sklearn.model_selection import train_test_split
```

```python
X, y = make_classification(n_samples=500, n_features=2, n_informative=2, n_redundant=0)
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
model = LogisticRegression().fit(X,y)
y_score = model.fit(X_train, y_train).decision_function(X_test)
y_pred = model.fit(X_train, y_train).predict(X_test)
```

```python
# calculate AUC
roc_auc = roc_auc_score(y_test, y_score)
# calculate roc curve
fpr, tpr, thresholds = roc_curve(y_test, y_score)
##false-positive rate, true positive rate, 임계값
```

```python
print(roc_auc)
##성능이 굉장히 좋다고 생각할 수 있다.
```

```python
plt.figure()
plt.plot(fpr, tpr, lw=2, color='darkorange', label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate = Recall')
plt.ylabel('True Positive Rate = Sensitivity')
plt.title('ROC curve')
plt.legend(loc="lower right")
plt.show()

##AUC도 높고 ROC 그래프도 왼쪽 위에 붙어있는 것으로 보아 좋은 성능을 가진 모델이라고 판단할 수 있겠다. 
```

```python
print(confusion_matrix(y_test, y_pred))
##혼동 행렬을 다시 확인해주면 TP, FN이 높은 것을 확인할 수 있겠다.
```

# 3. Cross Validation

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
cv = KFold(n_splits=3, shuffle=True, random_state=0)

for train_index, test_index in cv.split(X): ##데이터가 3등분이 된다. 500개가 166개씩 나눈다. 
    print(X[train_index], y[test_index])
    
# cv.split(X)는 X를 k-fold로 나눈 뒤 각 fold에서의 (train_index, test_index)를 반환하는 iterable
```

```python
kfold = KFold(n_splits=5, shuffle=True, random_state=0) # target이 골고루 분포되도록 shuffle 옵션
##5개의 fold로 실제 데이터를 나눠서 보도록 하자.
```

```python
scores = cross_val_score(LogisticRegression(), X, y, cv=kfold)  # cross_val_score(모델, 데이터, 타깃)
scores # 각 fold에 대한 score
```

```python
scores = cross_val_score(LogisticRegression(), X, y) 
scores 
# cross_val_score의 기본값은 k=3 괄호의 마지막을 비워두었을 때 기본값은 3이다. 
```

```python
scores = cross_val_score(LogisticRegression(), X, y, cv=5)
scores
# cross_val_score는 기본적으로 분류 모델에 대해 StratifiedKfold 사용
```

```python
display(scores.mean(), scores.var()) # 교차검증 점수들의 평균과 분산 확인
```

# 4. Grid Search

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
iris_data = load_iris()
```

```python
X_train, X_test, y_train,y_test = train_test_split(iris_data.data,iris_data.target, test_size = 0.2, random_state = 121)

##여기서는 test size를 0.2로 지정을 해준 상태이다. 
```

```python
dtree = DecisionTreeClassifier()
```

```python
parameters = {'max_depth':[1,2,3], 'min_samples_split':[2,3]}

##tree를 쳐 나갈때 몇단계까지 내려갈 수 있는지에 대한 것이 max_depth이고
##min_samples split이 몇개씩 나눠지는지이다. 
```

```python
grid_dtree = GridSearchCV(dtree, param_grid = parameters, cv = 3, refit = True)
```

```python
grid_dtree.fit(X_train, y_train)
```

```python
scores_df = pd.DataFrame(grid_dtree.cv_results_)
```

```python
scores_df[['params', 'mean_test_score','rank_test_score','split0_test_score','split1_test_score','split2_test_score']]

##밑의 표 해석 : mean test score는 평균 score이고 이 평균은 3 fold 교차 검증을 했는데 이거를 평균을 내준 것이다. 
## 높은 점수대로 등수를 메겨준다. 
##최고의 parameter를 찾아내고 그 값이 0.975라는 것도 알 수 있다. 
```

```python
print('GridSearchCV 최적 파라미터: ', grid_dtree.best_params_)
print('GridSearchCV 최고 정확도: {0:.4f}'.format(grid_dtree.best_score_))

##둘다 최적인데 앞에 있는 것을 가져와서 출력을 해준 것 같다. 
```

```python
estimator = grid_dtree.best_estimator_

pred = estimator.predict(X_test)
print('테스트 데이터 세트 정확도: {0:.4f}'.format(accuracy_score(y_test,pred)))

##test data를 마지막으로 돌려준 것이다.
```
